```python
import os

file_path = 'destination_folder/Resumes/'

# Initialize lists to store file names and folder names
file_names = []
folder_names = []

# Iterate through files in the specified directory
for file in os.listdir(file_path):
    # Check file extension and append file name and folder name to respective lists
    if file.endswith('.doc') or file.endswith('.docx'):
        file_names.append(file)
        folder_names.append('Resumes')

#you can print each file name along with its corresponding folder name
for file_name, folder_name in zip(file_names, folder_names):
    print("File:", file_name, "Folder:", folder_name)
```

```python
import os
import pandas as pd

# Specify the path to the main directory containing subfolders with resumes
main_folder_path = 'destination_folder/Resumes/'

# Initialize lists to store subfolder names and file names
subfolder_names = []
file_names = []

# Process files in the main directory first
main_folder_files = [file for file in os.listdir(main_folder_path) if file.endswith('.doc') or file.endswith('.docx')]
subfolder_names.extend(['Resumes'] * len(main_folder_files))
file_names.extend(main_folder_files)

# Iterate through the main directory and its subfolders
for root, dirs, files in os.walk(main_folder_path):
    for subfolder in dirs:
        subfolder_path = os.path.join(root, subfolder)
        subfolder_files = [file for file in os.listdir(subfolder_path) if file.endswith('.doc') or file.endswith('.docx')]
        subfolder_names.extend([subfolder] * len(subfolder_files))
        file_names.extend(subfolder_files)

# Create a DataFrame with subfolder names and file names
df = pd.DataFrame({'Profile': subfolder_names, 'Resumes': file_names})

# Adjust display options to show all rows, columns, and avoid column wrapping
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Display the DataFrame
print(df)

```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords

# Ensure NLTK stopwords are downloaded
try:
    nltk.data.find('corpora/stopwords.zip')
except LookupError:
    nltk.download('stopwords')

# Assuming 'resume_data' is your DataFrame containing resume details
# Make sure 'Resume_Details' column contains the text data
# Handle missing values if necessary

# Initialize CountVectorizer
countvec = CountVectorizer(stop_words=stopwords.words('english'), ngram_range=(1, 2))

# Fit and transform the text data into a matrix of n-grams
ngrams = countvec.fit_transform(resume_data['Resume_Details'])

# Count frequency of n-grams
count_values = ngrams.toarray().sum(axis=0)

# Retrieve vocabulary
vocab = countvec.vocabulary_

# Create DataFrame with n-gram frequencies
df_ngram = pd.DataFrame(sorted([(count_values[i], k) for k, i in vocab.items()], reverse=True),
                        columns=['Frequency', 'Ngram'])

# You can now use df_ngram for further analysis or visualization
```

```python

fig, axe = plt.subplots(1,1, figsize=(12,6), dpi=200)
ax = sns.barplot(x=df_ngram['Unigram_Bigram'].head(25), y=df_ngram.Frequency.head(25), data=resume_data, ax = axe,
            label='Total Pofile Category : {}'.format(len(resume_data.Category.unique())))

axe.set_xlabel('Words', size=16,fontweight= 'bold')
axe.set_ylabel('Frequency', size=16, fontweight= 'bold')
plt.xticks(rotation = 90)
plt.legend(loc='best', fontsize= 'x-large')
plt.title('Top 25 Most used Words in Resumes', fontsize= 18, fontweight= 'bold')

for i in ax.containers:
    ax.bar_label(i,color = 'black', fontweight = 'bold', fontsize= 12)

pylab.rcParams.update(rcParams)
fig.tight_layout()
plt.show()
fig.savefig('IMG/Top_Words_Bar', dpi = 500)
```
